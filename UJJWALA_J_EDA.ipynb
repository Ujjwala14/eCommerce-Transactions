```python
# Importing necessary libraries
import pandas as pd
import numpy as np
```

```python
# Loading the datasets
customers_df = pd.read_csv("Customers.csv")
products_df = pd.read_csv("Products.csv")
transactions_df = pd.read_csv("Transactions.csv")
```

```python
# Basic Overview of the Data
print("Customers Dataset:")
print(customers_df.info())
print(customers_df.head(), "\n")
```

```python
print("Products Dataset:")
print(products_df.info())
print(products_df.head(), "\n")
```

```python
print("Transactions Dataset:")
print(transactions_df.info())
print(transactions_df.head(), "\n")
```

```python
# Checking for missing values
print("Missing Values in Customers Dataset:")
print(customers_df.isnull().sum(), "\n")
```

```python
print("Missing Values in Products Dataset:")
print(products_df.isnull().sum(), "\n")
```

```python
print("Missing Values in Transactions Dataset:")
print(transactions_df.isnull().sum(), "\n")
```

```python
# Basic statistics for numeric columns
print("Statistics for Transactions Dataset:")
print(transactions_df.describe(), "\n")
```

```python
# Unique values for categorical columns
print("Unique Regions in Customers Dataset:")
print(customers_df['Region'].unique(), "\n")
```

```python
print("Unique Product Categories in Products Dataset:")
print(products_df['Category'].unique(), "\n")
```

```python
# Merging the datasets for further analysis
merged_df = pd.merge(transactions_df, customers_df, on='CustomerID', how='inner')
merged_df = pd.merge(merged_df, products_df, on='ProductID', how='inner')
```

```python
print("Merged Dataset:")
print(merged_df.info())
print(merged_df.head(), "\n")
```

```python
# Selecting only numeric columns for correlation analysis
numeric_columns = merged_df.select_dtypes(include=[np.number])
```

```python
# Computing and displaying the correlation matrix
print("Correlation Matrix:")
print(numeric_columns.corr(), "\n")
```

```python
# Heatmap for the correlation matrix
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix for Numeric Columns")
plt.show()
```

```python
# Data Cleaning: Checking for duplicates
print("Number of duplicate rows in merged dataset:")
print(merged_df.duplicated().sum(), "\n")
```

```python
# Top Customers by Total Spending
top_customers = merged_df.groupby('CustomerName')['TotalValue'].sum().sort_values(ascending=False)
print("Top 5 Customers by Spending:")
print(top_customers.head(), "\n")
```

```python
# Top Products by Total Sales
top_products = merged_df.groupby('ProductName')['TotalValue'].sum().sort_values(ascending=False)
print("Top 5 Products by Sales:")
print(top_products.head(), "\n")
```

```python
# Transactions over time
merged_df['TransactionDate'] = pd.to_datetime(merged_df['TransactionDate'])
sales_over_time = merged_df.groupby('TransactionDate')['TotalValue'].sum()
```

```python
# Plot: Sales over time
plt.figure(figsize=(12, 6))
sales_over_time.plot()
plt.title("Sales Over Time")
plt.xlabel("Transaction Date")
plt.ylabel("Total Sales (USD)")
plt.grid()
plt.show()
```

```python
# Plot: Region-wise sales
region_sales = merged_df.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales (USD)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot: Product category distribution
plt.figure(figsize=(10, 6))
products_df['Category'].value_counts().plot(kind='bar', color='salmon')
plt.title("Product Category Distribution")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.xticks(rotation=45)
plt.show()
```

```python
# Average order value by region
avg_order_value_by_region = merged_df.groupby('Region')['TotalValue'].mean().sort_values(ascending=False)
print("Average Order Value by Region:")
print(avg_order_value_by_region, "\n")
```

```python
# Product sales distribution
plt.figure(figsize=(10, 6))
merged_df['ProductName'].value_counts().head(10).plot(kind='bar', color='green')
plt.title("Top 10 Products by Number of Transactions")
plt.xlabel("Product Name")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.show()
```
